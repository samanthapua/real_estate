Data Cleansing

1.Removing NA from RegionName

2.Remove when price is null

3.Drop when landsize is 0 or null

4.Bedroom2 if null follow Rooms

Replace type
H- house,cottage,villa,semi,terrace
U-unit,duplex
T- townouse

Replace method
Method:
S - property sold;
SP - property sold prior;
PI - property passed in;
PN - sold prior not disclosed;
SN - sold not disclosed;
NB - no bid;
VB - vendor bid;
W - withdrawn prior to auction;
SA - sold after auction;
SS - sold after auction price not disclosed.
N/A - price or highest bid not available.

```python
import pandas as pd


data = pd.read_csv('./Melbourne_housing_FULL.csv').dropna(subset=['Regionname','Price','Landsize'])
```

```python
#to check all are nulls are dropped
print(data['Regionname'].unique())
print(data['Price'].isna().unique()) 
print(data['Landsize'].isna().unique())
```

```python
data.count()
```

```python
data = data[(data['Landsize'] != 0)]
```

```python
data['Bedroom2'] = data['Bedroom2'].fillna(data['Rooms'])
data
```

```python
data['Type'] = data['Type'].replace('h', 'House, Cottage, Villa, Terrace')
data['Type'] = data['Type'].replace('u', 'Unit, Duplex')
data['Type'] = data['Type'].replace('t', 'Town House')
data
```

```python
data['Method'] = data['Method'].replace('S','Property sold')
data['Method'] = data['Method'].replace('SP','Property sold prior')
data['Method'] = data['Method'].replace('PI','Property passed in')
data['Method'] = data['Method'].replace('PN','Property sold prior not disclosed')
data['Method'] = data['Method'].replace('SN','Property sold not disclosed')
data['Method'] = data['Method'].replace('NB','No bid')
data['Method'] = data['Method'].replace('VB','Vendor bid')
data['Method'] = data['Method'].replace('W','Withdrawn prior to auction')
data['Method'] = data['Method'].replace('SA','Sold after auction')
data['Method'] = data['Method'].replace('SS','Sold after auction price not disclosed')
data['Method'] = data['Method'].replace('N/A','Price or highest bid not available')

data
```

```python
data.to_csv('Melbourne_housing_FULL.csv',index=False)
```
